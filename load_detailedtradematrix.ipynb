More reasonnable number of item groups, selection by hand become possible.

```python
def items_in_group(item_group,df_item_groups):
    return df_item_groups[df_item_groups["Item Group"]==item_group]["Item"].unique()
```

```python
items_in_group("Meat of Swine",item_groups_product)
```

```python
items_in_group("Bovine Meat",item_groups_product)
```

```python
def select_item_from_group(df_to_filter,item_group,df_item_groups):
    # return dataframe with rows from df_to_filter having "Item" in item_group
    # definition of item_group is obtained from df_item_groups
    return df_to_filter[df_to_filter["Item"].isin(items_in_group(item_group,df_item_groups))]
```

```python
import numpy as np

n_rows_import = []
n_countries_reporter_import = []
n_countries_partner_import = []
n_countries_import = []

n_rows_export = []
n_countries_reporter_export = []
n_countries_partner_export = []
n_countries_export = []

for year in years :
    n_rows_import.append(len(import_q[import_q[year].notna()]))
    n_countries_reporter_import.append(len(import_q[import_q[year].notna()]["Reporter Countries"].unique()))
    n_countries_partner_import.append(len(import_q[import_q[year].notna()]["Partner Countries"].unique()))
    reporters_import = import_q[import_q[year].notna()]["Reporter Countries"].unique()
    partners_import = import_q[import_q[year].notna()]["Partner Countries"].unique()
    n_countries_import.append(len(np.unique(np.concatenate((reporters_import,partners_import)))))
    
    n_rows_export.append(len(export_q[export_q[year].notna()]))
    n_countries_reporter_export.append(len(export_q[export_q[year].notna()]["Reporter Countries"].unique()))
    n_countries_partner_export.append(len(export_q[export_q[year].notna()]["Partner Countries"].unique()))
    reporters_export = export_q[export_q[year].notna()]["Reporter Countries"].unique()
    partners_export = export_q[export_q[year].notna()]["Partner Countries"].unique()
    n_countries_export.append(len(np.unique(np.concatenate((reporters_export,partners_export)))))
```

```python
merged_trade_matrix = (exploded_import.merge(exploded_export,
                      on=["Importer Code","Importer","Exporter Code","Exporter","Item Code","Item","Year"],
                      how="outer",
                      suffixes=(' Importer Report', ' Exporter Report'))
                       [["Importer Code","Importer","Exporter Code","Exporter","Item Code","Item","Year","Unit Importer Report","Quantity Importer Report","Unit Exporter Report","Quantity Exporter Report"]])
```

```python
def get_value(importer, exporter, year, item, priority, merged_trade_matrix):
    """
    priority = {"import", "export"}
    Information taken in priority from either exporter or importer country report
    """
    #more efficient than standard access
    row = merged_trade_matrix.query(
        "(Importer == @importer) & (Exporter == @exporter) & (Year == @year) & (Item == @item)"
    )
    if priority == "import":
        if row["Quantity Importer Report"].isna().any():
            result = (row[["Unit Exporter Report", "Quantity Exporter Report"]])
        else:
            result = (row[["Unit Importer Report", "Quantity Importer Report"]])
    else:
        if row["Quantity Exporter Report"].isna().any():
            result = (row[["Unit Importer Report", "Quantity Importer Report"]])
        else:
            result = (row[["Unit Exporter Report", "Quantity Exporter Report"]])
    if len(result):
        return tuple(result.iloc[0])
    return np.nan, np.nan
```

```python
get_value("Algeria", "Australia", "Y1986", "Milk, whole dried", "export", merged_trade_matrix)
```
